### 算数（数学）

整数の特性は頻繁に登場します

- 量子力学は積分は使わず、合計だけが行われています
    - 合計だけとは積分に対応してであり、数式は色々あります
    - 合計とはExcelとかプログラミングのsum
    - 最先端物理は時間も必要ないらしいです
    - 工学的な積分は近似値と言えます。精度さえ取れれば、近似でも誤り訂正（のようなもの）をすれば問題ありません
- 超準解析は実数で無限小を表現しているように見えました。実数を使うこと自体が矛盾でこの理論が生まれたのでは？
- 有理数は遅延評価がなければ、現実(量子世界)では限定的にしか使えないように見えますが、工学的には使えます
    - 計算の止まらない有理数はありえませんが、Haskellで採用された遅延評価であり得る可能性も考えられます
- 高度な数学や工学は頻繁に体験できなくても、依存度という意味では頻繁に依存しています
